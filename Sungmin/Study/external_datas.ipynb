```python
#중복 데이터 삭제
new_data=[]
obser=[] #관측소 이름 정하기
for value in data:
    if value['observatoryName'] in obser:
        continue
    else:
        new_data.append(value)
        obser.append(value['observatoryName'])
new_data
```
