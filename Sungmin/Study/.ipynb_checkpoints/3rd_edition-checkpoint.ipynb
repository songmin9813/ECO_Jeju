```python
import pandas as pd
import numpy as np
import datetime
```

```python
korean=pd.read_csv("C:/Users/mitha/OneDrive/바탕 화면/dev/ECO_Jeju/Sungmin/new_datas/2nd_edition/korean_group.csv")
long=pd.read_csv("C:/Users/mitha/OneDrive/바탕 화면/dev/ECO_Jeju/Sungmin/new_datas/2nd_edition/long_group.csv")
short=pd.read_csv("C:/Users/mitha/OneDrive/바탕 화면/dev/ECO_Jeju/Sungmin/new_datas/2nd_edition/short_group.csv")
card=pd.read_csv("C:/Users/mitha/OneDrive/바탕 화면/dev/ECO_Jeju/Sungmin/new_datas/2nd_edition/card_group.csv")
waste=pd.read_csv("C:/Users/mitha/OneDrive/바탕 화면/dev/ECO_Jeju/Sungmin/new_datas/2nd_edition/waste_group.csv")
resident=pd.read_csv("C:/Users/mitha/OneDrive/바탕 화면/dev/ECO_Jeju/Sungmin/new_datas/2nd_edition/resident_group.csv")
```

- 변수별 월단위 변수 생성/수정 후 재조립 과정
- 각 테이블 별 row의 수에 유의하도록 함

```python
resident['resident_base_month']=resident['resident_base_month'].apply(lambda x:str(x).zfill(2))#월을 두자리수로 변경
resident['resident_base_year']=resident['resident_base_year'].astype(str)#년도를 스트링타입으로 변경
resident['base_date']=resident['resident_base_year']+'-'+resident['resident_base_month']#새로운 기준일 생성
resident['emd_nm']=resident['resident_emd_nm']#컬럼 이름 변경 귀찮아서 생성 후 삭제
resident.drop(['resident_base_year','resident_base_month','resident_emd_nm','Unnamed: 0'],axis=1,inplace=True)#필요없는 변수 삭제
resident_group=resident.groupby(['base_date','emd_nm']).sum()#월단위 읍면동 그룹핑
resident_group
```

```python
card['base_date']=card['base_date'].apply(lambda x:x[:7])#일 단위 삭제
card_group=card.groupby(['base_date','emd_nm']).sum()#월단위 읍면동 그룹핑
card_group
```

```python
long['base_date']=long['base_date'].apply(lambda x:x[:7])#일 단위 삭제
long_group=long.groupby(['base_date','emd_nm']).sum()#월단위 읍면동 그룹핑
long_group
```

```python
korean['base_date']=korean['base_date'].apply(lambda x:x[:7])#일 단위 삭제
korean_group=korean.groupby(['base_date','emd_nm']).sum()#월단위 읍면동 그룹핑
korean_group
```

```python
short['base_date']=short['base_date'].apply(lambda x:x[:7])#일 단위 삭제
short_group=short.groupby(['base_date','emd_nm']).sum()#월단위 읍면동 그룹핑
short_group
```

```python
waste['base_date']=waste['base_date'].apply(lambda x:x[:7])#일 단위 삭제
waste_group=waste.groupby(['base_date','emd_nm']).sum()#월단위 읍면동 그룹핑
waste_group
```

- 수정 완료. 각 테이블을 저장하도록 함

```python
waste_group.to_csv("C:/Users/mitha/OneDrive/바탕 화면/dev/ECO_Jeju/Sungmin/new_datas/3rd_edition/waste_group.csv")
card_group.to_csv("C:/Users/mitha/OneDrive/바탕 화면/dev/ECO_Jeju/Sungmin/new_datas/3rd_edition/card_group.csv")
long_group.to_csv("C:/Users/mitha/OneDrive/바탕 화면/dev/ECO_Jeju/Sungmin/new_datas/3rd_edition/long_group.csv")
short_group.to_csv("C:/Users/mitha/OneDrive/바탕 화면/dev/ECO_Jeju/Sungmin/new_datas/3rd_edition/short_group.csv")
korean_group.to_csv("C:/Users/mitha/OneDrive/바탕 화면/dev/ECO_Jeju/Sungmin/new_datas/3rd_edition/korean_group.csv")
resident_group.to_csv("C:/Users/mitha/OneDrive/바탕 화면/dev/ECO_Jeju/Sungmin/new_datas/3rd_edition/resident_group.csv")
```
