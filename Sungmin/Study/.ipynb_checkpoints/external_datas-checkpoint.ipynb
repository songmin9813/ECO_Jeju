- 모든 재료 생성됨. 일단 관련 파일 저장

```python
weather=pd.read_csv("C:/Users/mitha/OneDrive/바탕 화면/dev/ECO_Jeju/Sungmin/new_datas/weather.csv")
weather['base_date']=weather['baseDate']#컬럼 이름 변경
weather.drop(['Unnamed: 0','observatoryName','baseDate'],axis=1,inplace=True)#관련없는 데이터 삭제
weather.dropna(inplace=True)#결측치 삭제
weather['base_date']=weather['base_date'].astype(str)#스트링 타입으로 변경
weather['base_date']=weather['base_date'].apply(lambda x:x[:4]+"-"+x[4:6])#기본 날짜를 통일화
weather_group=weather.groupby(['base_date','emd_nm']).mean()#월별 읍면동 평균 온도로 그룹화
weather_group.to_csv("C:/Users/mitha/OneDrive/바탕 화면/dev/ECO_Jeju/Sungmin/new_datas/weather_group.csv")#csv 파일로 저장
weather.head()
```
