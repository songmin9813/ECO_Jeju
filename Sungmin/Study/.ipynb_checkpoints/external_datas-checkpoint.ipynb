- 연습 끝. 실코딩 시작

```python
#teimdelta에서 제공하는 날짜 추가는 일단위까지만 지워하였기 때문에 월별로 넘기기 위한 전용 함수가 필요했다.
one=[1,3,5,7,8,10,12]#31일인 날
two=[4,6,9,11]#30일인 날
def date_set(month,year):
    if month in one:
        return 31
    elif month in two:
        return 30
    elif year%4==0:
        return 29
    else:
        return 28
```

```python
end_point=datetime.datetime(2021,7,1)#중단점 설정
start_point=datetime.datetime(2018,1,1)#시작점 설정
decode='LtnHMv0xzkWvmnGl/RL6KClaKtT8scEa1T2Mrl0nopr1hZnNaieTX8WY8jTumW6mT8eD6rJXHVelgnsgYhKvww=='
list_=[]#모든것을 담을 리스트
while end_point!=start_point:
    if(start_point==datetime.datetime(2021,4,1)):#2021년 4월 데이터 일단 제외
        start_point+=datetime.timedelta(days=date_set(start_point.month,start_point.year))
        continue
    endpoint1='http://apis.data.go.kr/B552584/RfidFoodWasteServiceNew/getCityDateList?ServiceKey={}&type=json&page=1&rowNum=100&disYear={}&disMonth={}&cityCode=W6X'.format(decode,start_point.year,str(start_point.month).zfill(2))
    endpoint2='http://apis.data.go.kr/B552584/RfidFoodWasteServiceNew/getCityDateList?ServiceKey={}&type=json&page=1&rowNum=100&disYear={}&disMonth={}&cityCode=W6Y'.format(decode,start_point.year,str(start_point.month).zfill(2))
    resp1 = requests.get(endpoint1)
    resp2 = requests.get(endpoint2)
    data1=resp1.json()['data']['list']
    data2=resp2.json()['data']['list']
    for value in data1:
        list_.append(value)
    for value in data2:
        list_.append(value)
    start_point+=datetime.timedelta(days=date_set(start_point.month,start_point.year))
df=pd.DataFrame(data=list_)
```

- 만들어진 데이터를 이용하여 전처리 과정(이름 통일, 월별 통일, 필요없는 변수 제거)
- 제주시와 서귀포시만을 기준으로 삼아지기 때문에 이를 나눌 근거도 찾으면 좋을 듯 하다.

```python
df=pd.read_csv("C:/Users/mitha/OneDrive/바탕 화면/dev/ECO_Jeju/Sungmin/new_datas/external/rfid.csv")
df['disYear']=df['disYear'].astype(str)
df['disMonth']=df['disMonth'].astype(str)#각 타입을 스트링 타입으로 변경
df['disMonth']=df['disMonth'].apply(lambda s:s.zfill(2))#월 자릿수 통일
df['base_date']=df['disYear']+'-'+df['disMonth']#연도와 월을 합친 기준변수 생성
df['emd_nm']=df['citySggName']#다른 기준 변수 이름 통일
df.drop(['Unnamed: 0','disDate','disMonth','disDay','cityCode','citySidoName','citySggName','disQuantityRate','disCountRate'],axis=1,inplace=True)#필요없는 변수 제거
```

```python
dfsum=df.groupby(['emd_nm','disYear']).sum()#년도별 합이 저장된 데이터프레임 퍼센티지 계산할 때 사용함
```

```python
rfid=df.groupby(['base_date','emd_nm']).sum()
rfid['yearQuantityRate']=rfid['disQuantity']
rfid['yearQuantityRate']=rfid['yearQuantityRate'].apply(lambda x: x/dfsum.loc[rfid.index[rfid['yearQuantityRate']==x][0][1],rfid.index[rfid['yearQuantityRate']==x][0][0][:4]]['disQuantity']*100)
#개쩌는 공식1
rfid['yearCountRate']=rfid['disCount']
rfid['yearCountRate']=rfid['yearCountRate'].apply(lambda x: x/dfsum.loc[rfid.index[rfid['yearCountRate']==x][0][1],rfid.index[rfid['yearCountRate']==x][0][0][:4]]['disCount']*100)
#솔직히 좀 쩔었다
```

```python
import pandas as pd
import numpy as np
import pickle
import requests
import datetime

decode='NSBgec4lQ8x7Jc%2ByKwX%2BbNZgISv4RceHXTgZE8ywS9%2FtuIxp%2FviPdwahJ1t7SZ7G8YI6syKNP35HGotJKY%2FdXA%3D%3D' #api 주소값
inpath="C:/Users/mitha/OneDrive/바탕 화면/su1st.csv" #파일 불러오기 경로. su1st나 je1st 둘 중에 하나만 사용
#su1st : 서귀포시-682 row 존재 , je1st : 제주시-1913개 row 존재
outpath="C:/Users/mitha/OneDrive/바탕 화면/su235.csv" #데이터프레임 저장 경로/결과물 저장 경로
start=0
end=None #235단위로 끊어서
#end값이 row값 넘어가게 되면 인덱스 에러 발생. 470~682 범위에서는 None 요망

sudf=pd.read_csv(inpath)
sudf.drop(['Unnamed: 0'],axis=1,inplace=True)

#teimdelta에서 제공하는 날짜 추가는 일단위까지만 지워하였기 때문에 월별로 넘기기 위한 전용 함수가 필요했다.
one=[1,3,5,7,8,10,12]#31일인 날
two=[4,6,9,11]#30일인 날
def date_set(month,year):
    if month in one:
        return 31
    elif month in two:
        return 30
    elif year%4==0:
        return 29
    else:
        return 28
    
suApt=list(sudf['aptCode'])
suEmd=list(sudf['emd_nm'])
suApt=suApt[start:end]
suEmd=suEmd[start:end]
sulist=[]

for i,code in enumerate(suApt):
    end_point=datetime.datetime(2021,7,1)#중단점 설정
    start_point=datetime.datetime(2018,1,1)#시작점 설정
    while end_point!=start_point:
        endpoint='http://apis.data.go.kr/B552584/RfidFoodWasteServiceNew/getCityAptDateList?ServiceKey={}&type=json&page=1&rowNum=100&disYear={}&disMonth={}&cityCode=W6Y&aptCode={}'.format(decode,start_point.year,str(start_point.month).zfill(2),code)
        resp = requests.get(endpoint)
        if resp.json()['data']['resultCode'] != 'OK':#불러오기 오류
            print(str(start_point.year)+'년 '+str(start_point.month).zfill(2)+'월의 '+code+'자료가 불러와지지 않음')
            sulist.append({'base_date':str(start_point.year)+'-'+str(start_point.month).zfill(2),'aptCode':code,'emd_nm':suEmd[i]})
        else:#불러오기 완료
            print(str(start_point.year)+'년 '+str(start_point.month).zfill(2)+'월의 '+code+'자료')
            data=resp.json()['data']['list']
            disCountsum=0
            disQuansum=0
            for value in data:
                disQuansum+=value['disQuantity']
                disCountsum+=value['disCount']
            sulist.append({'base_date':str(start_point.year)+'-'+str(start_point.month).zfill(2),'aptCode':code,'emd_nm':suEmd[i],'disQuantity':disQuansum,'disCount':disCountsum})
        start_point+=datetime.timedelta(days=date_set(start_point.month,start_point.year))
        
df_new_su=pd.DataFrame(data=sulist)
df_new_su.to_csv(outpath)
```
